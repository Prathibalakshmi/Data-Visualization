### Linearly separable data

```python
center = [[0,0],[5,5]]
radius = 3
n = 200

r = 0 + (radius-0)*np.random.random(size=n)
theta1 = np.random.uniform(-np.pi,np.pi,n)
theta2 = np.random.uniform(-np.pi,np.pi,n)

x1 = center[0][0] + r*np.cos(theta1)
y1 = center[0][1] + r*np.sin(theta1)
data1 = np.hstack((np.matrix(x1).T,np.matrix(y1).T))

x2 = center[1][0] + r*np.cos(theta2)
y2 = center[1][1] + r*np.sin(theta2)
data2 = np.hstack((np.matrix(x2).T,np.matrix(y2).T))

data = np.vstack((data1,data2))
```

```python
plt.figure(figsize=(9,7))
plt.scatter(np.ravel(data1[:,0]),np.ravel(data1[:,1]),color='red',s=10,label='class1')
plt.scatter(np.ravel(data2[:,0]),np.ravel(data2[:,1]),color='blue',s=10,label='class2')
plt.title("Separable Data")
plt.legend()
plt.show()
```

### Non-linearly separable data

```python
def gen_points_in_ring(center, r_inner, r_outer, n=100):
    theta = np.linspace(-np.pi,np.pi,n)
    r = np.random.uniform(r_inner, r_outer, n)

    x = center[0] + r*np.cos(theta)
    y = center[1] + r*np.sin(theta)
    return np.hstack((np.matrix(x).T, np.matrix(y).T))
```

```python
center = [0,0]
data1 = gen_points_in_ring(center=center, r_inner=0.2, r_outer=0.3, n=500)
data2 = gen_points_in_ring(center=center, r_inner=0.4, r_outer=0.5, n=500)
plt.figure(figsize=(9,7))
plt.scatter(np.ravel(data1[:,0]),np.ravel(data1[:,1]),color='red',s=10,label='class1')
plt.scatter(np.ravel(data2[:,0]),np.ravel(data2[:,1]),color='blue',s=10,label='class2')
plt.title("Non-linearly separable Data")
plt.legend()
plt.show()
```

```python
data1 = np.hstack((data1,np.matrix(np.zeros(data1.shape[0])).T))
data2 = np.hstack((data2,np.matrix(np.ones(data2.shape[0])).T))

data = np.vstack((data1,data2))
```

## PCA on Yale dataset

```python
dir_path = '../week8/YALE/'
file_ls = os.listdir(dir_path)
```

```python
file_ls[0]
```

### Reading files from folder

```python
file = os.path.join(dir_path,file_ls[0])
img = plt.imread(file)
plt.figure(figsize=(6,8))
plt.imshow(img,cmap='gray')
plt.title(file_ls[0])
plt.axis('off')
plt.show()
```

```python
idx_list = [str(i).zfill(2) for i in range(1,16,1)]
idx_list
```

### Train set and Test set segregation

```python
## segregate the images per subject
subjects = [[] for i in range(15)]
for k,id in enumerate(idx_list):
    subject_name = "subject"+id
    for file in file_ls:
        if file.startswith(subject_name):
            subjects[k].append(file)
```

```python
for sub in subjects[0]:
    img_path = os.path.join(dir_path,sub)
    img = plt.imread(img_path)
    plt.figure(figsize=(6,8))
    plt.imshow(img,cmap='gray')
    plt.title(sub)
    plt.axis('off')
    plt.show()
```

```python
## train test split
test = []
train = []
for subject_ls in subjects:
    rand_img = np.random.choice(subject_ls)
    test.append(rand_img)
    for sub in subject_ls:
        if sub != rand_img:
            train.append(sub)
```

```python
print(len(train),len(test))
```

```python
## Reading train set into a train matrix
train_matr = []
for img in train:
    img_path = os.path.join(dir_path,img)
    image = plt.imread(img_path)
    row, col = image.shape
    image = image.reshape((row*col))
    train_matr.append(image)
train_matr = np.array(train_matr)
```

```python
mu = np.mean(train_matr,axis=0)
C = np.cov(train_matr)
```

```python
eigvalue, eigvector = np.linalg.eig(C)
```

### Scree plot

```python
plt.figure(figsize=(9,7))
plt.plot(eigvalue)
plt.title("Eigenvalues of Yale Dataset")
plt.show()
```

```python
sort_idx = np.argsort(eigvalue)
sort_idx  = sort_idx [::-1]
eigvalue = eigvalue[sort_idx]
eigvector = eigvector[sort_idx]
```

```python
total = np.sum(eigvalue)
summ = 0
k = 0
while summ/total < 0.90:
    summ += eigvalue[k]
    k += 1
```

```python
print(k)
```

```python
best_eigenvalue = eigvalue[:k]
best_eigenvector = eigvector[:k]
```

```python
train_matr.shape
```

### Eigen Faces

```python
best_eigenvector_ls = []
for i in range(k):
    eig = train_matr.T@best_eigenvector[i,:]
    eig = eig/best_eigenvalue[i]
    best_eigenvector_ls.append(np.ravel(eig))
```

```python
for eig in best_eigenvector_ls:
    comp_img = eig.reshape(row,col)
    comp_img = comp_img.astype(float)
    plt.imshow(comp_img, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
Q = np.array(best_eigenvector_ls).T
Q.shape
```

```python
mu.shape
```

### Feature vectors of testing set

```python
feat_vec_ls = []
for file in test:
    img_path = os.path.join(dir_path, file)
    img = plt.imread(img_path)
    row,col = img.shape
    img = img.reshape(row*col)
    feat_vec = Q.T@(img - mu)
    feat_vec_ls.append(feat_vec)
```

```python
len(feat_vec_ls[0])
```

```python
for i in range(len(test)):
    plt.stem(feat_vec_ls[i])
    plt.title("Feature vector of {}".format(test[i]))
    plt.show()
```
